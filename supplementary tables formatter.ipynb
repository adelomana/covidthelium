```python
import pandas
```

```python
input_file = '/Users/adrian/research/covithelium/results/module_annotation/pathways_table_all_modules_APR2024_NEW.csv'
input_file = '/Users/adrian/research/covithelium/results/module_annotation/pathways_table_all_modules_MAY2024.csv'
```

```python
df = pandas.read_csv(input_file, index_col=7)
df.drop(labels=['Module ID', 'Unnamed: 0'], axis=1, inplace=True)
df.index.rename('Module ID', inplace=True)
df.rename(columns={'No. annotations in module':'Number of features', 'Total no.features in module':'Number of annotated features'}, inplace=True)
df.head()
```

```python
# deal with features
translate_table = {ord("'") : None, 
               ord('[') : None, 
               ord(']') : None}

formatted_features = []
for feature in df['Representative feature(s)']:
    formatted_names = []
    feature = feature.replace('\n', '')
    names = feature.split("' '")
    for name in names:
        new = name.translate(translate_table)
        formatted_names.append(new)
    formatted_names_string = ", ".join(formatted_names)
    formatted_features.append(formatted_names_string)

df['Annotations for representative feature'] = formatted_features
df.head()
```

```python
# deal with enriched pathways
translate_table = {ord("'") : None, 
               ord('[') : None, 
               ord(']') : None}

formatted_features = []
for feature in df['Pathways']:
    #print(feature)
    formatted_names = []
    feature = feature.replace('\n', '')
    names = feature.split("' '")
    for name in names:
        new = name.translate(translate_table)
        formatted_names.append(new)
    formatted_names_string = ", ".join(formatted_names)
    formatted_features.append(formatted_names_string)
    #print(formatted_names_string)
    #print()

df['Enriched pathways'] = formatted_features
df.head()
```

```python
# deal with P values
translate_table = {ord("'") : None, 
               ord('[') : None, 
               ord(']') : None}

formatted_features = []
for feature in df['Pathway p.values']:
    print(feature)
    formatted_names = []
    feature = feature.replace('\n', '')
    names = feature.split()
    for name in names:
        new = name.translate(translate_table)
        formatted_names.append(new)
    formatted_names_string = ", ".join(formatted_names)
    formatted_features.append(formatted_names_string)
    print(formatted_names_string)
    print()

df['Enrichments P'] = formatted_features
df.head()
```

```python
# reorder
storage = df.loc[:, ['Number of features in module', 
           'Annotations for representative feature', 
           'Enriched pathways', 
           'Enrichments P']]
storage.head()
```
